```python
#part a
landscape= [5, 8, 6, 12, 9, 7, 17, 14, 10, 6, 19, 15, 11, 8]
```

```python
#first choice function
def first_choice_hc(landscape, start):
    current = start
    path = [current]

    while True:
        current_val = landscape[current - 1]
        moved = False

#we will check left side first
        if current > 1:
            left = current - 1
            if landscape[left - 1] > current_val:
                current = left
                path.append(current)
                moved = True
                continue

        if current < len(landscape):
            right = current + 1
            if landscape[right - 1] > current_val:
                current = right
                path.append(current)
                moved = True
                continue


        if not moved:
            break

    return current, path
```

```python
#stochastic hill climbing
import random

def stoch_hc(landscape, start):
  current = start
  path = [current]

  while True:
    high_list = []
    current_val = landscape[current - 1]



    if current > 1:
      left = current - 1
      if landscape[left - 1] > current_val:
        high_list.append(left)



    if current < len (landscape):
      right = current + 1
      if landscape[right - 1] > current_val:
        high_list.append(right)


    if not high_list:
      break

    else:
      neighbour = random.choice(high_list)
      current = neighbour
      path.append(current)

  return current,path
```

```python
import random
import math

#Run hill climbing multiple times from random starting points
#Keep track of the best solution found overall

def random_restart_hc(landscape, num_restarts, variant='first choice'):
  best_state = None
  best_value = -math.inf
  all_results = []

  for restart in range (num_restarts):
    start = random.randint(0, len(landscape) - 1)
    if variant == 'first choice':
      terminal, path = first_choice_hc(landscape, start)
    else:
      terminal, path = stoch_hc(landscape, start)

    value = landscape[terminal]
    all_results.append((start, terminal, path))

    if value > best_value:
      best_value = value
      best_state = terminal
  return (best_state, best_value, all_results)
```

```python
def main():

    global_max_value = max(landscape)
    num_restarts = 20

    for variant in ['first choice', 'stochastic']:
        print(f"\nRunning RRHC with variant: {variant}")
        print("Restart\tStart\tTerminal\tF-value\tGlobal Max?")

        best_state, best_value, all_results = random_restart_hc(landscape, num_restarts, variant)

        global_count = 0

        for i, (start, terminal, path) in enumerate(all_results):
            f_value = landscape[terminal]
            if f_value == global_max_value:
                is_global = 'Yes'
                global_count += 1
            else:
                is_global = 'No'

            print(f"{i+1}\t{start}\t{terminal}\t\t{f_value}\t{is_global}")


        discovery_rate = (global_count / num_restarts) * 100
        print(f"Global Maximum discovery rate: {discovery_rate:.2f}")

if __name__ == "__main__":
    main()
```

```python
#write a helper find local maxima(landscape) that
#returns all positions whose f-value exceeds both
#neighbours. Print all local maxima for the Q2 landscape.

def find_local_maxima(landscape):
  #at every point of landscape[i] check if it itself is greater then its neighbours
  maxima = []
  for i in range (1, len(landscape)-1):
    if landscape[i] > landscape[i-1] and landscape[i] > landscape[i+1]:
      maxima.append(i+1)
  return maxima

maxima = find_local_maxima(landscape)
print("Local maxima positions:", maxima)
```

```python
import random
import math

#Run hill climbing multiple times from random starting points
#Keep track of the best solution found overall

def random_restart_hc_plateau(landscape, num_restarts, variant='first choice'):
  best_state = None
  best_value = -math.inf
  all_results = []
  plateau_counter = 0

  for restart in range (num_restarts):
    start = random.randint(0, len(landscape) - 1)
    if variant == 'first choice':
      terminal, path = first_choice_hc(landscape, start)
    else:
      terminal, path = stoch_hc(landscape, start)

    value = landscape[terminal]
    if terminal in [6,7]:
      plateau_counter += 1
    all_results.append((start, terminal, path))

    if value > best_value:
      best_value = value
      best_state = terminal
  return (best_state, best_value, all_results, plateau_counter)
```
