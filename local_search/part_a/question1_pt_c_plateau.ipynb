```python
landscape = [4, 9, 6, 11, 15, 15, 15, 7, 13, 5, 16, 12]
```

```python
import random


#first choice with plateau detection
def fc_plateau(landscape, start):
    current = start
    path = [current]
    plat = False

    while True:
        current_val = landscape[current-1]
        moved = False
        equal_list = []

        #check left
        if current > 1:
            if landscape[current-2] > current_val:
                current = current - 1
                path.append(current)
                moved = True
                continue
            elif landscape[current-2] == current_val:
                equal_list.append(current-1)

        #check right
        if current < len(landscape):
            if landscape[current] > current_val:
                current = current + 1
                path.append(current)
                moved = True
                continue
            elif landscape[current] == current_val:
                equal_list.append(current+1)

        #plateau condition
        if not moved:
            if equal_list:
                plat = True
                print("Plateau encountered at state", current)
            break

    return plat


#stochastic with plateau detection
def stoch_plateau(landscape, start):
    current = start
    path = [current]
    p_count = 0

    while True:
        current_val = landscape[current-1]
        high_list = []
        equal_list = []

        #left
        if current > 1:
            if landscape[current-2] > current_val:
                high_list.append(current-1)
            elif landscape[current-2] == current_val:
                equal_list.append(current-1)

        #right
        if current < len(landscape):
            if landscape[current] > current_val:
                high_list.append(current+1)
            elif landscape[current] == current_val:
                equal_list.append(current+1)

        #move to better
        if high_list:
            current = random.choice(high_list)
            path.append(current)

        #plateau stop
        else:
            if equal_list:
                p_count += 1
                print("Plateau encountered at state", current)
            break

    return p_count


#first choice with sideways moves
def fc_sideways(landscape, start):
    current = start
    path = [current]
    sideways = 0

    while True:
        current_val = landscape[current-1]
        moved = False
        equal_list = []

        #left
        if current > 1:
            if landscape[current-2] > current_val:
                current = current - 1
                path.append(current)
                moved = True
                continue
            elif landscape[current-2] == current_val:
                equal_list.append(current-1)

        #right
        if current < len(landscape):
            if landscape[current] > current_val:
                current = current + 1
                path.append(current)
                moved = True
                continue
            elif landscape[current] == current_val:
                equal_list.append(current+1)

        #sideways move
        if not moved and sideways < 10 and equal_list:
            current = random.choice(equal_list)
            path.append(current)
            sideways += 1
            moved = True

        if not moved:
            break

    return path, current


#stochastic with sideways moves
def stoch_sideways(landscape, start):
    current = start
    path = [current]
    sideways = 0

    while True:
        current_val = landscape[current-1]
        high_list = []
        equal_list = []

        #left
        if current > 1:
            if landscape[current-2] > current_val:
                high_list.append(current-1)
            elif landscape[current-2] == current_val:
                equal_list.append(current-1)

        #right
        if current < len(landscape):
            if landscape[current] > current_val:
                high_list.append(current+1)
            elif landscape[current] == current_val:
                equal_list.append(current+1)

        #better move
        if high_list:
            current = random.choice(high_list)
            path.append(current)

        #sideways move
        elif sideways < 10 and equal_list:
            current = random.choice(equal_list)
            path.append(current)
            sideways += 1

        else:
            break

    return path, current

def main():

    p_count = 0
    print("\nFirst Choice Plateau:")
    for start in range(1, 13):
        p = fc_plateau(landscape, start)
        if p:
            p_count += 1
    print("Total runs getting stuck on plateau:", p_count)
    print("success rate: ", round((12 - p_count) / 12 * 100,2), "%")

    p_count = 0
    print("\nStochastic Plateau:")
    for start in range(1, 13):
        p = stoch_plateau(landscape, start)
        if p > 0:
            p_count += 1
    print("Total runs getting stuck on plateau:", p_count)


    global_max = max(landscape)

    print("\nSideways moves")
    print(f"{'Start':<6} {'Algorithm':<22} {'Path':<35} {'Term':<6} {'Steps'}")

    fc_success = 0
    for start in range(1, 13):
        path, term = fc_sideways(landscape, start)
        if landscape[term-1] == global_max:
            fc_success += 1
        print(f"{start:<6} FC-Sideways           {str(path):<35} {term:<6} {len(path)-1}")
    print(f"FC Sideways Success Rate: {fc_success / 12:.2%}")

    stoch_success = 0
    for start in range(1, 13):
        path, term = stoch_sideways(landscape, start)
        if landscape[term-1] == global_max:
            stoch_success += 1
        print(f"{start:<6} Stoch-Sideways        {str(path):<35} {term:<6} {len(path)-1}")
    print(f"Stoch Sideways Success Rate: {stoch_success / 12:.2%}")

if __name__ == "__main__":
    main()
```
