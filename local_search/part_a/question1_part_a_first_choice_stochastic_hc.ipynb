```python
landscape = [4, 9, 6, 11, 8, 15, 10, 7, 13, 5, 16, 12]
```

```python
#first choice function
def first_choice_hc(landscape, start):
    current = start
    path = [current]

    while True:
        current_val = landscape[current - 1]
        moved = False

#we will check left side first
        if current > 1:
            left = current - 1
            if landscape[left - 1] > current_val:
                current = left
                path.append(current)
                moved = True
                continue

        if current < len(landscape):
            right = current + 1
            if landscape[right - 1] > current_val:
                current = right
                path.append(current)
                moved = True
                continue


        if not moved:
            break

    return path, current
```

```python
#stochastic hill climbing
import random

def stoch_hc(landscape, start):
  current = start
  path = [current]

  while True:
    current_val = landscape[current - 1]
    high_list = []


    if current > 1:
      left = current - 1
      if landscape[left - 1] > current_val:
        high_list.append(left)



    if current < len (landscape):
      right = current + 1
      if landscape[right - 1] > current_val:
        high_list.append(right)


    if not high_list:
      break

    else:
      neighbour = random.choice(high_list)
      current = neighbour
      path.append(current)

  return path, current
```

```python
def main():
  print("Start  Algorithm       Path                Terminal State   Steps")

  for start in range(1,13):
    path_fc, term_fc = first_choice_hc (landscape, start)
    steps_fc = len (path_fc)-1
    print(f"{start:<6} First-Choice    {str(path_fc):<19} {term_fc:<16} {steps_fc}")

  for start in range(1,13):
      path_sch, term_sch = stoch_hc (landscape, start)
      steps_sch = len (path_sch)-1
      print(f"{start:<6} Stochastic      {str(path_fc):<19} {term_fc:<16} {steps_fc}")

if __name__ == "__main__":
    main()
```
