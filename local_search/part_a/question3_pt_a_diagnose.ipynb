```python
def diagnose_hc(land, start):
    curr = start
    path = []

    while True:
        if curr in path:
            idx = curr - 1
            print(f"Terminated at state {curr} with f={land[idx]}. Failure mode: ridge")
            return

        path.append(curr)
        idx = curr - 1
        val = land[idx]

        #move to better neighbor
        if idx > 0 and land[idx - 1] > val:
            curr = curr - 1
            continue
        if idx < len(land) - 1 and land[idx + 1] > val:
            curr = curr + 1
            continue

        #check for flat neighbors (equal values)
        flatL = (idx > 0 and land[idx - 1] == val)
        flatR = (idx < len(land) - 1 and land[idx + 1] == val)

        #no better or equal neighbors means its local max
        if not flatL and not flatR:
            print(f"Terminated at state {curr} with f={val}. Failure mode: local maximum")
            return

        #decide next state
        if flatL:
            nextState = curr - 1
        else:
            nextState = curr + 1

        #if revisiting then ridge
        if nextState in path:
            print(f"Terminated at state {curr} with f={val}. Failure mode: ridge")
        else:
            print(f"Terminated at state {curr} with f={val}. Failure mode: plateau")
        return


def main():
    #landscape 1: Local Max
    print("Testing Landscape 1 (Local Max): [3, 8, 4, 1, 1]")
    diagnose_hc([3, 8, 4, 1, 1], 2)

    #landscape 2: Plateau
    print("\nTesting Landscape 2 (Plateau): [2, 5, 5, 5, 2]")
    diagnose_hc([2, 5, 5, 5, 2], 2)

    #landscape 3: Ridge
    print("\nTesting Landscape 3 (Ridge / Oscillation): [1, 7, 7, 7, 1]")
    diagnose_hc([1, 7, 7, 7, 1], 2)


if __name__ == "__main__":
    main()
```
