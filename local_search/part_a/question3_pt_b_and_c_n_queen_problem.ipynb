```python
import random
def generate_random_board():
    board = list(range(8))
    random.shuffle(board)
    return board

board = [2, 4, 3, 1, 5, 0, 7, 6]
print (board)
```

```python
def count_conflicts(board):
    conflicts = 0

    for i in range(8):
        for j in range(i+1, 8):

            if abs(board[i] - board[j]) == abs(i - j):
                conflicts += 1
                print("conflict at", (board[i], i), "and", (board[j], j))

    print("conflicts:", conflicts)
    return conflicts

count_conflicts(board)
```

```python
import random

def stochastic_hc_nqueens(board):
    current = board
    current_conflicts = count_conflicts(current)

    attempts = 0
    max_attempts = 100

    while attempts < max_attempts:

        # goal check
        if current_conflicts == 0:
            return current

        # try random swap
        i, j = random.sample(range(8), 2)
        new_board = current.copy()
        new_board[i], new_board[j] = new_board[j], new_board[i]

        new_conflicts = count_conflicts(new_board)

        #accept if better
        if new_conflicts < current_conflicts:
            current = new_board
            current_conflicts = new_conflicts
            attempts = 0
        else:
            attempts += 1
            print(attempts)

    #stuck so return best found
    return current
stochastic_hc_nqueens(board)
```

```python
import random

def solve_nqueens_rrhc(num_restarts):

    for restart in range(num_restarts):

        #generate random board
        board = list(range(8))
        random.shuffle(board)

        print(f"Restart {restart}: starting board = {board}")

        #run stochastic HC
        result = stochastic_hc_nqueens(board)


        conflicts = count_conflicts(result)

        print(f"End of restart {restart}, conflicts = {conflicts}")

        #check solution
        if conflicts == 0:
            print("Solved at restart:", restart)
            return result

    return "No solution found"

solve_nqueens_rrhc(100)
```

```python
def solve_nqueens_rrhc(num_restarts):
    for restart in range(num_restarts):
        #generate random board
        board = list(range(8))
        random.shuffle(board)

        #run stochastic HC
        result = stochastic_hc_nqueens(board)

        conflicts = count_conflicts(result)

        #check solution
        if conflicts == 0:
            return restart + 1  #number of restarts used
    return None  #no solution found
```

```python
k = [5, 10, 25, 50, 100]  # board sizes
trials = 30
results = []

for i in k:
    successes = 0
    restarts_list = []
    for _ in range(trials):
        restarts_needed = solve_nqueens_rrhc(100)  # max 100 restarts
        if restarts_needed is not None:
            successes += 1
            restarts_list.append(restarts_needed)
    success_rate = successes / trials
    avg_restarts = sum(restarts_list)/len(restarts_list) if restarts_list else 0
    results.append((k, success_rate, avg_restarts))
    print(f"k={k}, Success Rate={success_rate:.2f}, Avg Restarts={avg_restarts:.2f}")
```

```python
print(f"{'k':<6}{'Success Rate':<15}{'Avg Restarts'}")
for k, sr, ar in results:
    print(f"{k:<6}{sr:<15.2f}{ar:.2f}")
```
