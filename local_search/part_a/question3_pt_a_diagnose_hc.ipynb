```python
def diagnose_hc(landscape, start):
    current = start
    path = [current]
    visited = set()  # to detect ridge
    n = len(landscape)

    while True:
        moved = False
        current_val = landscape[current - 1]

        # ridge detection: revisiting a state
        if current in visited:
            print(f"Terminated at state {current} with f={current_val}. Failure mode: ridge")
            return
        visited.add(current)

        # check left neighbor
        if current > 1:
            left_val = landscape[current - 2]
            if left_val > current_val:
                current = current - 1
                path.append(current)
                moved = True
                continue
            elif left_val == current_val:
                # MOVE sideways to left
                current = current - 1
                path.append(current)
                moved = True
                continue

        # check right neighbor
        if current < n:
            right_val = landscape[current]
            if right_val > current_val:
                current = current + 1
                path.append(current)
                moved = True
                continue
            elif right_val == current_val:
                # MOVE sideways to right
                current = current + 1
                path.append(current)
                moved = True
                continue

        # if we have no better or equal move
        if not moved:
            mode = "local maximum"
            print(f"Terminated at state {current} with f={current_val}. Failure mode: {mode}")
            return
```

```python
landscape_r = [3,  5,  4 , 6]
landscape_p = [1,4,4,4,3,2]
landscape_lm = [2,3,1,6,5,3]
```

```python

print("Local Maximum")
diagnose_hc(landscape_lm, start=2)

print("\nPlateau")
diagnose_hc(landscape_p, start=3)

print("\nRidge")
diagnose_hc(landscape_r, start=2)
```

```python
def diagnose_hc(landscape, start):
    current     = start
    visited_set = set()
    on_plateau  = False

    while True:

        # ── FAILURE MODE 3 — RIDGE ──
        if current in visited_set:
            val = landscape[current][0]
            print(f"Terminated at state {current} with f={val}. Failure mode: ridge")
            return current, "ridge"

        visited_set.add(current)
        val  = landscape[current][0]
        nbrs = landscape[current][1]

        # ── First-Choice: take first strictly better neighbour ──
        moved_up = False
        for n in nbrs:
            if landscape[n][0] > val:
                current    = n
                on_plateau = False
                moved_up   = True
                break
        if moved_up:
            continue

        # ── No strictly better neighbour ──
        equal_nbrs = [n for n in nbrs if landscape[n][0] == val]

        if not equal_nbrs:
            # ── FAILURE MODE 1 or 2 ──
            if on_plateau:
                print(f"Terminated at state {current} with f={val}. Failure mode: plateau")
                return current, "plateau"
            else:
                print(f"Terminated at state {current} with f={val}. Failure mode: local maximum")
                return current, "local maximum"

        # ── Take first-choice equal step (plateau walk) ──
        on_plateau = True
        current    = equal_nbrs[0]
```

```python
# Local Maximum: A → B → C (peak)
landscape_local_max = {
    "A": (3, ["B"]),
    "B": (6, ["A", "C"]),
    "C": (10, ["B", "D", "E"]),
    "D": (7, ["C", "E"]),
    "E": (5, ["D", "C"])
}

# Plateau: P1 → P2 → P3 → P4
landscape_plateau = {
    "P1": (3, ["P2"]),
    "P2": (8, ["P1", "P3"]),
    "P3": (8, ["P4"]),
    "P4": (8, ["P5"]),
    "P5": (2, [])
}


landscape_ridge = {
    "R1": (1, ["R2"]),
    "R2": (5, ["R1", "R3"]),
    "R3": (9, ["R2", "R4"]),
    "R4": (9, ["R3", "R5"]),
    "R5": (3, ["R4"])
}
```

```python
print("=== Local Maximum ===")
diagnose_hc(landscape_local_max, start="A")

print("\n=== Plateau ===")
diagnose_hc(landscape_plateau, start="P1")

print("\n=== Ridge ===")
diagnose_hc(landscape_ridge, start="R1")
```
