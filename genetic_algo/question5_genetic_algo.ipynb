```python
import random
def random_chromosome():
  chr = []
  for i in range(6):
    chromosome = (random.randint(0,2), random.randint(0,3))
    chr.append(chromosome)
  return chr

#generates a random chromosome within range
chr = random_chromosome()
print (chr)
```

```python
def count_conflicts(chromosome):
  #avoids conflicts eg same room and same time slot
  conflicts = 0
  for i in range(len(chromosome)):
       for j in range(i+1, len(chromosome)):
           if chromosome[i] == chromosome[j]: #compares a course with itself
           #for example compares and reports 0,1 and 0,1
                conflicts += 1
  return conflicts

count_conflicts(chr)
```

```python
def fitness(chromosome):
    conflicts = count_conflicts(chromosome)
    return 100 - (10 * conflicts)

fitness (chr)
```

```python
#Generate and print 5 random chromosomes, their conflict counts, and fitness values

for i in range (5):
  chr = random_chromosome()
  count = count_conflicts(chr)
  fit = fitness(chr)
  print ("\nchromosome", i+1, ":", chr, "\nconflicts: ", count, "fitness: ", fit)
```

```python
def crossover(p1, p2, point):

    child = p1[:point] + p2[point:] #a part of p1 combined with a part of p2
    return child
```

```python
#a post-crossover repair function that detects conflicting courses and
#reassigns them to a random conflict-free (room, slot) combination.
#if no conflict-free slot exists, assign randomly.

def repair(chromosome):

    used = set()
    for i in range(len(chromosome)):
        #we check if this gene conflicts with any previously assigned
        while chromosome[i] in used:
            #we wil find a random new room and slot
            chromosome[i] = (random.randint(0,2), random.randint(0,3))
        used.add(chromosome[i])
    return chromosome
```

```python
def mutate(chromosome, pm=0.1):

    for i in range(len(chromosome)):
        if random.random() < pm:
            chromosome[i] = (random.randint(0,2), random.randint(0,3))
    return chromosome

#mutate according to mutation rate (0.1 in this case)
```

```python
chromo = [(1, 3), (1, 3), (1, 3), (2, 3), (2, 2), (0, 1)]
c=count_conflicts(chromo)
print("conflicts before:" , c)

r = repair (chromo)
rep = count_conflicts(r)
print ("conflicts after:", rep )
```

```python
def tournament_selection(pop):
    #pick two random indices
    i = random.randint(0, len(pop)-1)
    j = random.randint(0, len(pop)-1)

    #make sure they are different
    while j == i:
        j = random.randint(0, len(pop)-1)

    #return the better
    if fitness(pop[i]) >= fitness(pop[j]):
        return pop[i]
    else:
        return pop[j]
```

```python
def run_scheduling_ga(pop_size, generations, pm):
    #initialize population
    population = [random_chromosome() for i in range(pop_size)]
    best_fitness_per_gen = []

    for gen in range(generations):
        new_population = []

        while len(new_population) < pop_size:
            #selection
            parent1 = tournament_selection(population)
            parent2 = tournament_selection(population)

            #crossover
            child = crossover(parent1, parent2, 2)

            #repairr
            child = repair(child)

            #mutation
            child = mutate(child, pm)

            #add to new population
            new_population.append(child)

        #replace old population
        population = new_population

        #find best fitness
        # Find best chromosome in population
        best_chrom = max(population, key=fitness)


        best_fitness = fitness(best_chrom)

        if count_conflicts(best_chrom) == 0:
          print(f"Solution found at generation {gen}: {best_chrom}")
          break

    #after all generations, find best schedule
    best_chromosome = max(population, key=fitness)
    print("\nBest schedule found:")
    print(best_chromosome)
    print("Conflict count:", count_conflicts(best_chromosome))
    print("Fitness:", fitness(best_chromosome))

    return best_fitness

#a population of course schedules using tournament selection, crossover,
#repair, and mutation, showing the best fitness per generation
#repairing so everything is conflict-free, and printing the first
#conflict-free schedule

run_scheduling_ga(20, 50, 0.1)
```
